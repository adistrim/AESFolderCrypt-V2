```python
import matplotlib.pyplot as plt
import networkx as nx

steps = [
    "Input text",
    "Concatenate text and Salt",
    "Shuffle Characters",
    "XOR Hashing",
    "Repeat for 10 Iterations",
    "Output"
]

connections = [
    (0, 1),
    (1, 2),
    (2, 3),
    (3, 4),
    (4, 5),
    (4, 2)
]

# Create a directed graph
G = nx.DiGraph()

# Add nodes
for i, step in enumerate(steps):
    G.add_node(i, label=step)

# Add edges
G.add_edges_from(connections)

# Plot the graph with spring layout
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=33)
labels = {i: G.nodes[i]['label'] for i in G.nodes()}
nx.draw(G, pos, labels=labels, with_labels=True, node_size=2000, node_color='skyblue', font_size=10, font_weight='bold', arrowsize=20)
plt.title('Hashing Algorithm Flowchart', fontsize=14)
plt.savefig('hashing_algorithm.png', format='png')  # Save as .png file
plt.show()
```
